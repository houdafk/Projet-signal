### Tracer les 100 premiers échantillons et calculer à partir du graphe la fréquence du signal :
Notre étude se pose sur des sons purs ( un signal qui a une amplitude et une fréquence constante et qui par conséquent, varie au fonction du temps seulement car la forme du signal est : 

f(t)=A.Sin(2πft+φ) ). 

On a Fe=44100Hz.

Par définition la fréquence d'échantillonnage (Fe) est le nombre d'échantillons prit en une seconde (1000 ms), donc pour savoir le temps nécessaire pour prendre 100 échantillons on multiplie 100 par 1000 et on divise le tout par 44100 ( (100*1000)/44100 ) et qui va donner un résultats de 2,26ms.

Et donc pour tracer les 100 premiers échantillons il suffit de prendre notre fréquence choisie est dessiner son graphe de 0 à 2,26ms. 


```python
import sounddevice as sd
import wave
import math
import numpy as np
import matplotlib.pyplot as plot
nomFichier = 'son.wav' # Nom du fichier à créer
monSon = wave.open( nomFichier, 'w') # Ouverture fichier en écriture
nbCanal = 1 # Son monophonique (un seul canal)
nbOctet = 1 # Chaque échantillon est quantifié sur 8 bits
fech = 44100 # On échantillone à 44,1 kHz
nbEchantillons = 44100
parametres = (nbCanal, nbOctet, fech, nbEchantillons, 'NONE', 'not compressed')
monSon.setparams(parametres)
fch = 440
for i in range(nbEchantillons):
 val = wave.struct.pack('B',int( (255/2)*(1+math.sin(2*math.pi*fch*i/fech))))
 monSon.writeframes(val)
monSon.close()

n = np.linspace(0,1,fech)
x = np.sin(2*np.pi*fch*n)
plot.figure(figsize=(10,2))
plot.plot(n[:100],x[:100])
plot.show() 


sd.play(x,fech) #Jouer le son

```

Et voici le schéma du filtre :
 
- Donner l'impédance équivalente du circuit proposé, étudier les variations de son module en fonction de ω (pulsation), Déduire la fréquence de résonance du circuit et son coefficient de qualité Q :

 $$Zeq = \frac{JRCW-LCW^2+1}{JCW}$$

- c) la fréquence de résonance du circuit : 

$$f = \frac{W}{2π}= 23000$$

- d) coefficient de qualité Q :

$$Q = \frac{1}{R}*\sqrt{\frac{L}{C}}= \frac{1}{\sqrt{2}} $$

- Justifier les valeurs des composants pour une meilleure détection du signal non audible :
- a) La fréquence f : c’est celle qu’on a choisi au début pour le signale non audible quelle est à f = 23000Hz.

- b) l'inductance de la bobine L : on a pris une bobine idéale (r=0) et on a posé une valeur pour l’inductance L = 10mH.


- c) La capacité du condensateur C : d’après la loi de la pulsation :

$$ω=2πf=\frac{1}{\sqrt{LC}}$$ 

$$LC = \frac{1}{4π^2 f^2}$$  

$$C = \frac{1}{4πL^2 f^2}$$  

$$C = \frac{1}{4* 10 * 10^-3*π^2* 23000^2}$$  

$$C = 4.6 nF $$  

- La résistance R : Afin de déterminer la valeur de la résistance on a utilisé la loi du facteur de qualité :

$$Q = \frac{1}{R}*\sqrt{\frac{L}{C}}= \frac{1}{\sqrt{2}} $$  

$$RQ = \sqrt{\frac{L}{C}}$$  

$$R = \sqrt{\frac{L}{C}}*\frac{1}{Q}$$  

$$R= \sqrt{\frac{10*10^-3}{4.6*10^(-9)}}*\sqrt{2}$$  

$$R = 2085 Ω$$  

- Établir l'équation différentielle de l’évolution de la tension aux bornes du condensateur :
On utilisant la loi des mailles :


$$E = Ur+Uc  + Ul = Ri + Uc + L \frac{di}{dt}$$

$$\frac{d^2Uc}{dt^2}+\frac{R}{L}*\frac{dUc}{dt}+\frac{1}{LC}*Uc=\frac{E}{LC}$$
