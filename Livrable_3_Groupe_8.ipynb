    - Dans le cas d’un texte court : 

    La chaine de transmission reste tel qu’elle car il s’agit d’un simple fichier de petite taille.

    - Dans le cas d’un fichier volumineux :

    - Dans le cas d’un son :

L’agent infiltré active le microphone et enregistre l’onde sonore permettant une connexion au réseau dans la salle de conférence, il utilisera son pad afin de moduler le signal et donc l’envoyer dans les hautes fréquences, puis grâce à la transformé de fourrier il le filtre, et envoie une trame PPP (point par point) qui utilise une transmission synchrone et l’envoie par réseau. L’agent responsable de la surveillance du réseau reçoit la trame, démodule le signal, utilise l’inverse de la transformé de fourrer et lit le son.

    On résume que dans le cas :
-	d’un texte court il n’y aura aucun impact sur la chaine de transmission. 
-	d’un fichier volumineux, plusieurs trames seront envoyés ce qui prendra plus de temps.
-	d’un son, la chaine de transmission ne sera pas impacté mais d’autres démarche précéderont l’envoie de la trame.

   ## c) Les étapes de modulation : 
   
   La représentation en modulation par impulsions et codage d'un signal se décompose en trois opérations : échantillonnage, quantification et codage.


   ### 3.1 L’échantillonnage :
	
    
   ### Introduction : 
   
   Un signal analogique varie d’une manière continue en fonction du temps. Lors de la numérisation d’un tel signal, il y a un échantillonnage de celui-ci dans le temps. Cela veut dire que l’on ne mesure les valeurs du signal que périodiquement. La fréquence d'échantillonnage fe (en hertz) se définit comme l'inverse de la durée entre deux échantillons successifs (période d’échantillonnage Te en secondes) : $$ fe={\frac{1}{Te}}$$ 
     Ainsi, la fréquence d'échantillonnage correspond au nombre d'échantillons prélevés chaque seconde.


   

  ### Théorème d'échantillonnage de Nyquist–Shannon : 
  
  Le choix de la fréquence d’échantillonnage est crucial afin de reproduire fidèlement le signal étudié. En effet, si celui-ci varie trop vite par rapport à Fe, la numérisation donnera un rendu incorrect du signal. C’est le cas pour le graphe de droite, ci-dessous, où l’échantillonnage donne une sinusoïde avec une fréquence plus faible qu’en réalité. On parle d’aliasing.
 
   
  
   Shannon dit pour que le signal puisse être entièrement reconstruit à partir des échantillons, il faut et il suffit que : La fréquence d'échantillonnage doit être strictement supérieure à deux fois la plus grande fréquence présente dans le spectre du signal continu : Fe≥2Fmax 
   
   Dans notre cas notre fréquence d'échantillonnage sera 46 000 ( 23 000 * 2) afin d'avoir un fichier moins volumineux et du coup une transmission plus rapide.

   ## 3.2 La quantification:
	
 ### Le principe de quantification et son but : 
 
 L’opération de quantification consiste à attribuer un nombre binaire à toute valeur prélevée au signal lors de l’échantillonnage et qui permet de transformer un signal continu à un signal discret. C’est le CAN (convertisseur analogique numérique) qui réalise cette opération.
     Chaque niveau de tension est codé sur p bits, chaque bit pouvant prendre deux valeurs (0 ou 1).Donc un convertisseur à p bits possède 2^p niveaux de quantification. Considérons un CAN à 4 bits, il n’y a donc que  2^4=16 valeurs possibles attribuables à toutes les valeurs prélevées lors de l’échantillonnage.
     L’opération se fait donc avec une perte d’information d’autant plus grande que p est petit.

 ## 3.3. Le codage (le code Manchester): 

Le codage Manchester est un codage synchrone, ce qui signifie que, outre les données à transmettre, les signaux transmis intègrent également l'horloge de synchronisation nécessaire à leur décodage. Il est utilisé dans les réseaux informatiques pour injecter sur le média physique (couche 1 du modèle OSI) les valeurs logiques correspondant au flux d'entrée.

 ### Le principe du codage :


   Les transitions du signal codé transmettent à la fois la valeur logique du bit (0 ou 1) et l'instant de son échantillonnage. Une transition intermédiaire est ajoutée lorsque des bits de même valeur se suivent.


   Toutefois le schéma du code diffère entre le côté émetteur et le côté récepteur, même si leur référence commune est le signal de synchronisation :


- du côté émetteur, le signal doit être stabilisé à la valeur du bit à transmettre au moment du top de l'horloge qui provoque la transition de synchronisation ;


- du côté récepteur, la transition marque le top d'échantillonnage de la valeur du bit : celui-ci est égal à la valeur qui précédait cet instant ou l'inverse de la nouvelle valeur qui suit la transition.


  Si un codage simple peut s'effectuer en plaçant temporellement la synchronisation au centre du bit comme illustré ci-dessous, ce qui donne un signal « haut » durant 50 % du temps, des émetteurs alimentés par piles peuvent réduire leur durée d'émission (état ON) en raccourcissant les périodes "hautes" avant ou après la transition de synchronisation, voire en ajoutant des transitions intercalaires.



  ### Utilisation:

Ethernet 10BASE5, 10BASE2, 10BASE-T, 10BASE-FL.

Liaisons radio à courte distance : télécommandes radio, stations météo domestiques, ...

Réseau luminaires DALI (IEC 62386)


  ### Avantages

Mise en œuvre simple, codage et décodage faciles, pas de composante continue (donc pas de perte de synchronisation sur les suites de symboles identiques).

Les problèmes habituellement rencontrés avec les codes tels que NRZ, NRZI ou Miller :

perte de synchronisation ;

ligne coupée (à cause d'une rafale de 0) ;

sensibilité aux parasites ;

affaiblissement du signal car moyenne non nulle ;

sont résolus par le codage Manchester en supprimant les suites de 0 ou de 1.

  La limite basse de la bande passante occupée n'est plus le continu comme dans les trois codes donnés en exemple. Elle vaut sensiblement la moitié de sa limite haute.


  ## II. Démodulation :
	
    
  ### a) Définition : 
  
  La démodulation consiste à extraire le signal porteur d'informations d'origine d'une onde porteuse.

  
  ### b) Type de démodulation ASK : 
	
    
   #### Définition et principe : 
   
   La modulation d'amplitude consiste à modifier l'amplitude de l'onde porteuse par le signal modulant. Pour réaliser ceci on utilise un montage multiplieur qui effectue le produit des tensions Um(t) et Up(t) :

$$Um(t)=U0+s(t)=U0+Umcos(2πf.t + φ1)$$


$$Up(t)=Up cos(2πF.t +φ2) $$




-  Démodulateur ASK : Le détecteur ASK synchrone se compose d'un détecteur de loi carrée, d'un filtre passe-bas, d'un comparateur et d'un limiteur de tension. Voici le schéma de principe pour le même.


Le signal d'entrée modulé ASK est transmis au détecteur de loi Square. Un détecteur à loi carrée est un détecteur dont la tension de sortie est proportionnelle au carré de la tension d'entrée modulée en amplitude. Le filtre passe-bas minimise les fréquences les plus élevées. Le comparateur et le limiteur de tension aident à obtenir une sortie numérique propre.

   ### c) Etapes de démodulation : 
   
   Le but de la démodulation est de pouvoir récupérer le signal modulant (l'information) lors de la réception du signal modulé. . Il est donc nécessaire de :

- Supprimer les alternances négatives.

- Supprimer la porteuse (c'est à dire détecter l'enveloppe).

- Supprimer la tension continue de décalage Uo.


   ## III. les éléments de protocole envisagés :
 
 La transmission de données passe par un ensemble de règles qui régissent les échanges de données ou le comportement collectif de processus ou d'ordinateurs en réseaux ou d'objets connectés.
 
 ### 01- Type de liaisons :

Dans notre cas nous utiliseront une liaison half-duplex : Les données sont transmises dans les deux sens mais non simultanément (exploitation bidirectionnelle à l’alternat).  Le récepteur reçoit les données de l’émetteur puis peut envoyer à son tour des informations. Ce qui est très pratique pour les accusés de réception entre nos deux agents.

 ### 02- Codage :

   Le codage est une opération qui permet de déterminer pour chaque caractère ou groupe de bits, une valeur décimal ou hexadécimale ( code ASCII, ISO8859..)
	
   Nous utiliserons une transmission série synchrone : Dans laquelle émetteur et récepteur sont cadencés à la même horloge. Le récepteur reçoit de façon continue (même lorsque aucun bit n'est transmis) les informations au rythme où l'émetteur les envoie. C'est pourquoi il est nécessaire qu'émetteur et récepteur soient cadencés à la même vitesse. De plus, des informations supplémentaires sont insérées afin de garantir l'absence d'erreurs lors de la transmission.
	
   Lors d'une transmission synchrone, les bits sont envoyés de façon successive sans séparation entre chaque caractère, il est donc nécessaire d'insérer des éléments de synchronisation, on parle alors de synchronisation au niveau caractère. On l’utilise car celle-ci est d’un cout moins élevé et est très utile pour les transmissions de longue distance.
	
 ### 03- Méthode d’envoi de données : 
	
   La couche liaison de données prend les paquets de la couche réseau et les encapsule dans des trames. Si la taille de la trame devient trop grande, le paquet peut être divisé en trames de petite taille. À l'extrémité du récepteur, la couche liaison de données capte les signaux du matériel et les assemble en trames. 

   Dans notre cas nous utiliseront une trame PPP car sa trame dispose de : 


-	Fanion : il est de 1 octet avec le motif de bits 01111110 et il permet de détecter le début et la fin d’une trame.


-	Adresse : 1 octet mis à 11111111 en cas de diffusion.


-	Contrôle : 1 octet défini sur une valeur constante de 11000000.


-	Protocole : 1 ou 2 octets qui définissent le type de données contenues dans le champ de charge utile.


-	Charge utile : elle transporte les données de la couche réseau. La longueur maximale du champ de charge utile est de 1500 octets.

  Et surtout elle dispose du FSC (séquence de contrôle de trame de 2 ou 4 octets pour la détection d'erreur. Le code standard utilisé est CRC (code de redondance cyclique)) et elle envoie un accusé de réception automatiquement.

  PS: Les données envoyées dans une trame sont sous forme de code binaire.
  
  ### 04- Le type de données traitées à chaque étape :
  
  Pour un message ou un fichier volumineux : Numérique (pad) puis binaire (trame) puis Numérique a nouveau (oridnateur de l'agents responsable de la surveillance réseau).
  
  Pour un son : Analogique puis numérique (pad ) puis binaire (trame) puis numérique aprés analogique de nouveau (oridnateur de l'agents responsable de la surveillance réseau ).
